## Inference - bert for violence/non-violence + catboost for passive/active

### Stage 1 - Violence or not?


```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoConfig

model_name = "csebuetnlp/banglabert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
import datasets
import evaluate
import numpy as np
import pandas as pd
```

```python
config = AutoConfig.from_pretrained(
    model_name,
    num_labels=2,
    finetuning_task=None,
    cache_dir=None,
    revision="main",
    use_auth_token=None,
)

model_trained = AutoModelForSequenceClassification.from_pretrained("/content/drive/MyDrive/research/emnlp_blp_2023/task_1/models/banglabert_50_epochs_violence_binary.pt", config=config)
```

```python
test = pd.read_csv("/content/drive/MyDrive/research/emnlp_blp_2023/task_1/data/dev.csv")
test
```

```python
def modify_label(row):
  if row['label'] == 2:
    return 1
  else:
     return row['label']

test['violence'] = test.apply(lambda row : modify_label(row), axis=1)
test
```

```python
X_test = list(test["text"])
y_test = list(test["violence"])
# encodings = tokenizer(X_test, truncation=True, padding=True) #, max_length=128)

# test_dataset = VITD_Dataset(X_test, y_test)
# test_dataset
```

```python
from transformers import pipeline

classifier = pipeline("text-classification",tokenizer = tokenizer, model=model_trained)
```

```python
# test.shape[0]
from tqdm import tqdm
predictions = []
for i in tqdm(range(test.shape[0])):
  prediction = classifier(test["text"].iloc[i])
  if prediction[0]['label'] == 'LABEL_0':
    predictions.append(0)
  elif prediction[0]['label'] == 'LABEL_1':
    predictions.append(1)
  elif prediction[0]['label'] == 'LABEL_2':
    predictions.append(2)
```

```python
label = list(test["label"])
```

```python
d = {
    "text":X_test,
    "label":label,
    "predicted_1":predictions

}
```

```python
model_1_output = pd.DataFrame(d)
model_1_output
```

```python
model_1_output["label"].nunique()
```
